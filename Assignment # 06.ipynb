```python
Question: Define Object Oriented Programming Language?
    
Answer: Object-oriented programming (OOP) is a programming paradigm based on the concept of objects, which 
        can contain data in the form of fields (often known as attributes or properties), and code, in the 
        form of procedures (often known as methods.
```

```python
Question: List down the Benefits of OOP?

Answer: There are some benefits of OOP are listed below:
        1. Re-usability
        2. Data Redundancy
        3. Code Maintenance
        4. Security
        5. Design Benefits
        6. Better productivity
        7. Easy troubleshooting
        8. Flexibility through polymorphism
        9. Effective problem solving
```

```python
Question: Differentiate between function and method?
    
Answer:   Method:
            1. Method is called by its name, but it is associated to an object (dependent).
            2. A method is implicitly passed the object on which it is invoked.
            3. A method can operate on the data (instance variables) that is contained by the corresponding class.
            
          Function:
            1. Function is block of code that is also called by its name. (independent)
            2. The function can have different parameters or may not have any at all. If any data (parameters) are passed,
               they are passed explicitly.
            3. Function does not deal with Class and its instance concept.
            
```

```python
Question: Define the following terms:
            1. Class 
            2. Object 
            3. Attribute 
            4. Behavior
            
Answer:     Class:
                Classes are like a blueprint or a prototype that you can define to use to create objects.
                We define classes by using the class keyword, similar to how we define functions by using
                the def keyword.
                
            Object:
                An object (instance) is an instantiation of a class. When class is defined, only the description
                for the object is defined. Therefore, no memory or storage is allocated.
                
            Attribute:
                Classes contain characteristics called Attributes.
                
            Behavior:
                Behavior is one of the characteristics of an object.
```

```python
class Car():
    def __init__(car,name,model,color,engine):
        
        car.name = name
        car.model = model
        car.color = color
        car.engine = engine
        car.brand = "Toyota"
        
        
    def updateModel(car,newmodel):
        car.model += newmodel
        
    def updateColor(car,newcolor):
        car.color = newcolor
        
    def updateBrand(car,newbrand):
        car.brand = newbrand 
        
print("------------\n Car Detail \n------------")
        
Car1 = Car("Aqua",2018,"Black","1500CC")
print(Car1.brand)
print(Car1.name)
print(Car1.model)
print(Car1.color, "Old Color")
Car1.updateColor("White")
print(Car1.color, "New Color")
print(Car1.engine)
print("\n")

Car2 = Car("Vitz",2018,"Blue","1300CC")
print(Car2.brand)
print(Car2.name)
print(Car2.model)
print(Car2.color)
print(Car2.engine)
print("\n")

Car3 = Car("Prius",2018,"White","17980CC")
print(Car3.brand)
print(Car3.name)
print(Car3.model, "Old Model")
Car3.updateModel(1)
print(Car3.model, "New Model")
print(Car3.color)
print(Car3.engine)
print("\n")

Car4 = Car("Passo",2018,"Red","1300CC")
print(Car4.brand)
print(Car4.name)
print(Car4.model)
print(Car4.color)
print(Car4.engine)
print("\n")

Car5 = Car("Civic",2019,"Silver","1799CC")
print(Car5.brand, "Default Brand")
Car5.updateBrand("Honda")
print(Car5.brand, "Update Brand")
print(Car5.name)
print(Car5.model)
print(Car5.color)
print(Car5.engine)
print("\n")

print(Car1)
print(Car2)
print(Car3)
print(Car4)
print(Car5)
```
